# RealSense Camera Capture

This notebook walks through the `physicalai.capture` API for RealSense cameras:
discover devices, capture color and depth frames, and visualise aligned RGBD output.

**Prerequisites** — install the RealSense SDK bindings and matplotlib:

```bash
uv pip install physicalai[realsense] matplotlib
```

## 1. Discover available devices

````python
from physicalai.capture.cameras.realsense import RealSenseCamera

devices = RealSenseCamera.discover()
for i, dev in enumerate(devices):
    print(f"[{i}] {dev.name}  serial={dev.hardware_id}")

if not devices:
    raise RuntimeError("No RealSense devices found. Check USB connection.")
````

## 2. Connect

Pick the serial number from the list above.

````python
serial = devices[0].device_id  # change index if you have multiple cameras

cam = RealSenseCamera(serial_number=serial, fps=30, width=640, height=480)
cam.connect()
print(f"Connected to {cam.device_id}")
````

## 3. Capture a color frame

````python
import matplotlib.pyplot as plt

frame = cam.read()
print(f"shape={frame.data.shape}  dtype={frame.data.dtype}  seq={frame.sequence}")

plt.figure(figsize=(8, 6))
plt.imshow(frame.data)
plt.title("Color")
plt.axis("off")
plt.show()
````

## 4. Capture a depth frame

````python
depth = cam.read_depth()
print(f"shape={depth.data.shape}  dtype={depth.data.dtype}  seq={depth.sequence}")

plt.figure(figsize=(8, 6))
plt.imshow(depth.data, cmap="jet")
plt.colorbar(label="Depth (mm)")
plt.title("Depth")
plt.axis("off")
plt.show()
````

## 5. Aligned RGBD capture

`read_rgbd()` returns aligned color and depth from a single pipeline read.

````python
color, depth = cam.read_rgbd()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.imshow(color.data)
ax1.set_title(f"Color  seq={color.sequence}")
ax1.axis("off")
im = ax2.imshow(depth.data, cmap="jet")
ax2.set_title(f"Depth  seq={depth.sequence}")
ax2.axis("off")
fig.colorbar(im, ax=ax2, label="mm")
plt.tight_layout()
plt.show()
````

## 6. Disconnect

````python
cam.disconnect()
print(f"Connected: {cam.is_connected}")
````
